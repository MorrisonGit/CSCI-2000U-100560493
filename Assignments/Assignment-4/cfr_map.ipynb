```python
fig = plt.figure(figsize=(14,10))
ax = plt.subplot(1,1,1)
earth = Basemap(projection='mill')
earth.drawcoastlines(color='0.50', linewidth=0.25)
earth.fillcontinents(color='0.95')

for (name,lat,long,cases) in data:
    
    lo, la = map(float(long),float(lat))
    
    if name == "Measles":
        Measles = ax.scatter(lo, la,float(cases)*0.008,zorder=10,facecolors='none',edgecolors='red')

    if name == "Mumps":
        Mumps= ax.scatter(lo, la,float(cases)*0.0657,zorder=10,facecolors='none',edgecolors='green')
        
    if name == "Polio":
        Polio = ax.scatter(lo, la,float(cases)*2.5,zorder=10,facecolors='none',edgecolors='blue')
        
    if name == "Rubella":
        Rubella = ax.scatter(lo, la,float(cases)*0.0275,zorder=10,facecolors='none',edgecolors='magenta')


plot.legend((Measles,Mumps,Polio,Rubella),("Measles", "Mumps", "Polio", "Rubella"),
markerscale=10, loc='best', fontsize=25)

plot.show()
```
